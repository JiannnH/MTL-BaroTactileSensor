```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_t0 = pd.read_csv('../tip1\\test_full_generated.csv')
# df_t0 = pd.read_csv('../tip1\zv1_zd5_t8_20230622_181736.csv')

print(df_t0.shape)

# Plot the data
plt.figure(figsize=(10,5))
plt.scatter(df_t0['x'], df_t0['Sensor_0'], label='sensor0')
plt.scatter(df_t0['x'], df_t0['Sensor_1'], label='sensor1')
plt.scatter(df_t0['x'], df_t0['Sensor_2'], label='sensor2')
plt.scatter(df_t0['x'], df_t0['Sensor_3'], label='sensor3')
plt.legend()
plt.xlabel('x')
plt.ylabel('sensor value')
plt.title('sensor value vs x (before calibration)')
plt.grid(True)
plt.show()
```

```python
# Copy the first 1000 rows of data for Sensor_0
noisy_data_sensor_0 = df_t0['Sensor_0'].copy(deep=True).head(1000)

# Copy the last 1000 rows of data for other sensors
noisy_data_other_sensors = df_t0.iloc[-1000:, 1:].copy(deep=True)

# Add noise to the 'Sensor_0' column
noise_level = 0.2
noisy_data_sensor_0 += np.random.normal(0, noise_level, len(noisy_data_sensor_0))

# Add noise to the other sensor columns
noisy_data_other_sensors += np.random.normal(0, noise_level, noisy_data_other_sensors.shape)

# Combine the noisy data with the original data
df = pd.concat([df_t0, noisy_data_sensor_0, noisy_data_other_sensors], axis=1, ignore_index=True)
```

```python
# Define the amount of noise
noise_level = 0.2

# create a new dataframe to store the noisy data
noisy_data = pd.DataFrame()
# Copy the first 1000 rows of data

noisy_data['Sensor_0'] = df_t0['Sensor_0'].tail(1000).copy(deep=True)
noisy_data['Sensor_1'] = df_t0['Sensor_1'].tail(1000).copy(deep=True)
noisy_data['Sensor_2'] = df_t0['Sensor_2'].tail(1000).copy(deep=True)
noisy_data['Sensor_3'] = df_t0['Sensor_3'].tail(1000).copy(deep=True)


# Repeat the 1000 data 20 times
noisy_data = pd.concat([noisy_data]*40, ignore_index=True)




# Add noise to the 'Sensor_0' column
noisy_data['Sensor_0'] = noisy_data['Sensor_0'] + np.random.normal(0, noise_level, len(noisy_data))
noisy_data['Sensor_1'] = noisy_data['Sensor_1'] + np.random.normal(0, noise_level, len(noisy_data))
noisy_data['Sensor_2'] = noisy_data['Sensor_2'] + np.random.normal(0, noise_level, len(noisy_data))
noisy_data['Sensor_3'] = noisy_data['Sensor_3'] + np.random.normal(0, noise_level, len(noisy_data))

# Append the noisy data to the original data
df = pd.concat([df_t0, noisy_data], ignore_index=True)
```

```python
# plot the Sensor_0 data
plt.figure(figsize=(10,5))
plt.scatter(df.index, df['Sensor_0'], label='sensor0')
plt.scatter(df.index, df['Sensor_1'], label='sensor1')
plt.scatter(df.index, df['Sensor_2'], label='sensor2')
plt.scatter(df.index, df['Sensor_3'], label='sensor3')
plt.legend()
plt.xlabel('x')
plt.ylabel('sensor value')
plt.title('sensor value vs x (before calibration)')
plt.grid(True)
plt.show()
```

```python
mean = [df['Sensor_0'].iloc[:100].mean(), df['Sensor_1'].iloc[:100].mean(), df['Sensor_2'].iloc[:100].mean(), df['Sensor_3'].iloc[:100].mean(), df['Sensor_4'].iloc[:100].mean(), df['Sensor_5'].iloc[:100].mean(), df['Sensor_6'].iloc[:100].mean(), df['Sensor_7'].iloc[:100].mean(), df['Sensor_8'].iloc[:100].mean(), df['Sensor_9'].iloc[:100].mean(), df['Sensor_10'].iloc[:100].mean(), df['Sensor_11'].iloc[:100].mean(), df['Sensor_12'].iloc[:100].mean(), df['Sensor_13'].iloc[:100].mean(), df['Sensor_14'].iloc[:100].mean(), df['Sensor_15'].iloc[:100].mean()]
max = [df['Sensor_0'].max(), df['Sensor_1'].max(), df['Sensor_2'].max(), df['Sensor_3'].max(), df['Sensor_4'].max(), df['Sensor_5'].max(), df['Sensor_6'].max(), df['Sensor_7'].max(), df['Sensor_8'].max(), df['Sensor_9'].max(), df['Sensor_10'].max(), df['Sensor_11'].max(), df['Sensor_12'].max(), df['Sensor_13'].max(), df['Sensor_14'].max(), df['Sensor_15'].max()]
```

```python
# if sensor reading is smaller than mean, set it to mean
for i in range(0, 16):
    df['Sensor_'+str(i)][df['Sensor_'+str(i)] < mean[i]] = mean[i]

```

```python


df['Sensor_0'] = df['Sensor_0'] - mean[0]
df['Sensor_1'] = df['Sensor_1'] - mean[1]
df['Sensor_2'] = df['Sensor_2'] - mean[2]
df['Sensor_3'] = df['Sensor_3'] - mean[3]
df['Sensor_4'] = df['Sensor_4'] - mean[4]
df['Sensor_5'] = df['Sensor_5'] - mean[5]
df['Sensor_6'] = df['Sensor_6'] - mean[6]
df['Sensor_7'] = df['Sensor_7'] - mean[7]
df['Sensor_8'] = df['Sensor_8'] - mean[8]
df['Sensor_9'] = df['Sensor_9'] - mean[9]
df['Sensor_10'] = df['Sensor_10'] - mean[10]
df['Sensor_11'] = df['Sensor_11'] - mean[11]
df['Sensor_12'] = df['Sensor_12'] - mean[12]
df['Sensor_13'] = df['Sensor_13'] - mean[13]
df['Sensor_14'] = df['Sensor_14'] - mean[14]
df['Sensor_15'] = df['Sensor_15'] - mean[15]


# # Calibrate the sensor reading by devided the max value of each sensor
df['Sensor_0'] = df['Sensor_0'] / (max[0] - mean[0])
df['Sensor_1'] = df['Sensor_1'] / (max[1] - mean[1])
df['Sensor_2'] = df['Sensor_2'] / (max[2] - mean[2])
df['Sensor_3'] = df['Sensor_3'] / (max[3] - mean[3])
df['Sensor_4'] = df['Sensor_4'] / (max[4] - mean[4])
df['Sensor_5'] = df['Sensor_5'] / (max[5] - mean[5])
df['Sensor_6'] = df['Sensor_6'] / (max[6] - mean[6])
df['Sensor_7'] = df['Sensor_7'] / (max[7] - mean[7])
df['Sensor_8'] = df['Sensor_8'] / (max[8] - mean[8])
df['Sensor_9'] = df['Sensor_9'] / (max[9] - mean[9])
df['Sensor_10'] = df['Sensor_10'] / (max[10] - mean[10])
df['Sensor_11'] = df['Sensor_11'] / (max[11] - mean[11])
df['Sensor_12'] = df['Sensor_12'] / (max[12] - mean[12])
df['Sensor_13'] = df['Sensor_13'] / (max[13] - mean[13])
df['Sensor_14'] = df['Sensor_14'] / (max[14] - mean[14])
df['Sensor_15'] = df['Sensor_15'] / (max[15] - mean[15])

print(mean,max)
```
